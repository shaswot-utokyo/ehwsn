```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from common.env_lib import csv_solar_harvester
harvester = csv_solar_harvester(location='tokyo',
                                year=1995)
done = False
henergy_rec=[]
while not done:
    _,henergy,_,_,done = harvester.step()
    henergy_rec.append(henergy)
henergy_rec.append((0))
plt.hist(henergy_rec,
         bins=10,
         rwidth=0.95,
         align='mid')
plt.yscale('log')
plt.title("Histogram of Henergy at each timestep")
plt.show()

np_henergy_rec = np.array(henergy_rec)
print("Mean Henergy:", np_henergy_rec.mean())
print("Max Henergy", max(henergy_rec))
print("Min Henergy", min(henergy_rec))
np_henergy_rec = np_henergy_rec.reshape(-1,240)
daily_mean_henergy = np_henergy_rec.mean(axis=1)
plt.plot(daily_mean_henergy)
plt.title("Daily Henergy variation over a year")
plt.show()
print("Max daily Henergy", max(daily_mean_henergy))
print("Min daily Henergy", min(daily_mean_henergy))
hist = plt.hist(daily_mean_henergy,
                bins=10,
                rwidth=0.95, 
                align='mid')
plt.xticks(hist[1],rotation=45)
plt.title("Histogram of Daily Cumulative Henergy")
plt.show()
```
